```python
import os
import sys
import pandas as pd


REQUIRED_COLUMNS = {"order_date", "carrier", "delivery_time_days"}


def load_and_validate_csv(path: str) -> pd.DataFrame:
    # 1) Základní kontrola existence souboru
    if not os.path.exists(path):
        raise FileNotFoundError(f"Soubor '{path}' nebyl nalezen.")

    # 2) Načtení CSV (s jednoduchým ošetřením chyby čtení)
    try:
        df = pd.read_csv(path)
    except Exception as exc:
        raise RuntimeError(f"Nepodařilo se načíst CSV '{path}': {exc}") from exc

    # 3) Kontrola povinných sloupců
    missing = REQUIRED_COLUMNS - set(df.columns)
    if missing:
        raise ValueError(
            f"V CSV chybí povinné sloupce: {sorted(missing)}. "
            f"Nalezené sloupce: {list(df.columns)}"
        )

    # 4) Převod order_date na datetime + normalizace na den
    df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce").dt.normalize()
    if df["order_date"].isna().any():
        bad = df[df["order_date"].isna()].head(5)
        raise ValueError(
            "Některé hodnoty 'order_date' nelze převést na datum. Ukázka (max 5):\n"
            f"{bad}"
        )

    # 5) Převod delivery_time_days na číslo
    df["delivery_time_days"] = pd.to_numeric(df["delivery_time_days"], errors="coerce")
    if df["delivery_time_days"].isna().any():
        bad = df[df["delivery_time_days"].isna()].head(5)
        raise ValueError(
            "Některé hodnoty 'delivery_time_days' nelze převést na číslo. Ukázka (max 5):\n"
            f"{bad}"
        )

    return df


def compute_daily_and_baseline(df: pd.DataFrame) -> pd.DataFrame:
    # 6) Denní agregace: průměr delivery_time_days za (order_date, carrier)
    daily = (
        df.groupby(["order_date", "carrier"], as_index=False)["delivery_time_days"]
        .mean()
        .rename(columns={"delivery_time_days": "daily_avg_delivery"})
        .sort_values(["carrier", "order_date"])
        .reset_index(drop=True)
    )

    # 7) Pro každého dopravce doplnit chybějící dny + spočítat baseline
    parts = []
    for carrier_value, g in daily.groupby("carrier", sort=False):
        # 7a) Necháme jen to, co potřebujeme, a order_date dáme do indexu
        g = g.sort_values("order_date").set_index("order_date")[["daily_avg_delivery"]]

        # 7b) Kompletní denní osa pro dopravce (min..max)
        full_dates = pd.date_range(g.index.min(), g.index.max(), freq="D")

        # 7c) Reindex => doplnění chybějících dní (NaN v daily_avg_delivery)
        g = g.reindex(full_dates)
        g.index.name = "order_date"

        # 7d) Doplnit carrier konstantou
        g["carrier"] = carrier_value

        # 7e) Baseline: předchozích 14 kalendářních dní bez aktuálního dne, min_periods=7
        g["baseline_14d"] = (
            g["daily_avg_delivery"]
            .shift(1)
            .rolling(window=14, min_periods=7)
            .mean()
        )

        parts.append(g.reset_index())

    daily_full = pd.concat(parts, ignore_index=True)

    # 8) pct_change + anomálie
    # Vypočítat pct_change a příznak anomálie podle pravidla > 30 % nad baseline
    daily_full["pct_change"] = (daily_full["daily_avg_delivery"] / daily_full["baseline_14d"]) - 1
    # Anomálie počítáme jen tam, kde existuje aktuální hodnota i baseline
    daily_full["is_anomaly"] = (
        daily_full["daily_avg_delivery"].notna()
        & daily_full["baseline_14d"].notna()
        & (daily_full["daily_avg_delivery"] > 1.30 * daily_full["baseline_14d"])
    )

    # 9) Výstup ve formátu dle zadání
    return daily_full[
        ["order_date", "carrier", "daily_avg_delivery", "baseline_14d", "pct_change", "is_anomaly"]
    ].sort_values(["order_date", "carrier"]).reset_index(drop=True)


def main(input_path: str, output_path: str) -> None:
    # 10) Načíst a zvalidovat data
    df = load_and_validate_csv(input_path)

    # 11) Spočítat denní průměry + baseline + anomálie
    result = compute_daily_and_baseline(df)

    # 12) Vyfiltrovat anomálie a vypsat je
    anomalies = result[result["is_anomaly"]].copy()

    if anomalies.empty:
        print("Žádné anomálie nebyly nalezeny (daily_avg_delivery > 1.30 * baseline_14d).")
    else:
        print(anomalies.to_string(index=False))

    # 13) Uložit anomálie do CSV
    anomalies.to_csv(output_path, index=False)
    print(f"\nUloženo: {output_path}")


if __name__ == "__main__":
    try:
        main(
            r"C:\Users\Pavla\Downloads\alensa_dummy_dataset.csv",
            r"C:\Users\Pavla\Downloads\anomalies_delivery_time.csv",
        )
    except Exception as exc:
        print(f"Chyba: {exc}", file=sys.stderr)
        sys.exit(1)
```
